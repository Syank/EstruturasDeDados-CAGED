```python
dicionarioMeses = {
    1: "Janeiro", 
    2: "Fevereiro", 
    3: "Março", 
    4: "Abril", 
    5: "Maio", 
    6: "Junho", 
    7: "Julho", 
    8: "Agosto",
    9: "Setembro", 
    10: "Outubro", 
    11: "Novembro", 
    12: "Dezembro"
}
    
def pegarNomeMes(numeroMes):
    return dicionarioMeses[numeroMes]
```

```python
def lerTabelasDiretorio(ano, colunasUtilizaveis):
    caminhoDados = "../doc/Dados/CAGED"
    
    fonteDados = os.listdir(caminhoDados + "/" + ano)
    
    # Retira os arquivos que não são do tipo .txt da lista
    for arquivo in fonteDados:
        if ".txt" not in arquivo:
            del fonteDados[fonteDados.index(arquivo)]
            
    tabelas = {}
    
    for nomeTabela in fonteDados:
        valorMes = int(nomeTabela.split(".")[0][-2:])  # Pega os dois últimos valores do nome do arquivo, que representam o mês e converte para um inteiro
        nomeMes = pegarNomeMes(valorMes)
        tabelas[nomeMes] = pandas.read_csv(caminhoDados + "/" + ano + "/" + nomeTabela,
                                           delimiter=";",
                                           usecols=colunasUtilizaveis)
        
    return tabelas
```

```python
colunasUtilizaveis = ["sexo", "tipomovimentação", "município", "idade", "graudeinstrução"]

dados = lerTabelasDiretorio(anoDeConsulta, colunasUtilizaveis)
```

```python
def organizarPorRegiao(dadosCrus):
    novosDados = {}
    
    for mes in dadosCrus:
        for codigoRegiao in dicionarioDeDados["município"]:
            nomeMunicipio = dicionarioDeDados["município"][codigoRegiao]
            
            mascara = dadosCrus[mes]["município"] == codigoRegiao
            
            filtrado = dadosCrus[mes][mascara]
            
            if nomeMunicipio not in novosDados:
                novosDados[nomeMunicipio] = {
                    mes: filtrado
                }
            else:
                novosDados[nomeMunicipio][mes] = filtrado
    
    return novosDados
```

```python
def pegarSomaTotalGrupo(grupo, colunaParaSomar):
    return grupo.groupby(colunaParaSomar).count().values[0][0]

def pegarTipoColunaGrupo(grupo, nomeColuna):
    indexColuna = grupo.columns.get_loc(nomeColuna)
    return grupo.sample().values[0][indexColuna]
```

```python
def pegarSomaTotalGrupo(grupo, colunaParaSomar):
    return grupo.groupby(colunaParaSomar).count().values[0][0]

def pegarTipoColunaGrupo(grupo, nomeColuna):
    indexColuna = grupo.columns.get_loc(nomeColuna)
    return grupo.sample().values[0][indexColuna]

def agruparDados(dadosParaAgrupar):
    dadosAgrupados = {}
    
    for mes in dadosParaAgrupar:
        dadosAgrupados[mes] = {}
        
        for generoCru in dadosParaAgrupar[mes]:
            genero = dicionarioDeDados["sexo"][generoCru]
            
            if genero not in dadosAgrupados[mes]:
                dadosAgrupados[mes][genero] = {}
            
            for tipoMovimentacaoCru in dadosParaAgrupar[mes][generoCru]:
                tipoMovimentacao = dicionarioDeDados["tipomovimentação"][tipoMovimentacaoCru]
                
                if tipoMovimentacao not in dadosAgrupados[mes][genero]:
                    dadosAgrupados[mes][genero][tipoMovimentacao] = dadosParaAgrupar[mes][generoCru][tipoMovimentacaoCru]
                else:
                    dadosAgrupados[mes][genero][tipoMovimentacao] += dadosParaAgrupar[mes][generoCru][tipoMovimentacaoCru]
    
    return dadosAgrupados
            

def filtrarDados(dadosCrus):
    total = {}
    
    for mes in dadosCrus:
        grupo = dadosCrus[mes].groupby(["sexo", "tipomovimentação"])

        total[mes] = {}

        for chave, valor in grupo:
            grupinho = grupo.get_group(chave)

            totalMovi = pegarSomaTotalGrupo(grupinho, "tipomovimentação")
            tipoMovi = pegarTipoColunaGrupo(grupinho, "tipomovimentação")

            tipoSexo = pegarTipoColunaGrupo(grupinho, "sexo")

            if tipoSexo not in total[mes]:
                total[mes][tipoSexo] = {}

            if tipoMovi not in total[mes][tipoSexo]:
                total[mes][tipoSexo][tipoMovi] = totalMovi
            else:
                total[mes][tipoSexo][tipoMovi] += totalMovi
                
    dadosTratados = agruparDados(total)
    
    return dadosTratados

def organizarDados(dados):
    novosDados = {}
    for municipio in dadosPorRegiao:
        novosDados[municipio] = {}
        for mes in dadosPorRegiao[municipio]:
            grupos = dadosPorRegiao[municipio][mes].groupby(["graudeinstrução", "tipomovimentação"])

            dadosGrupos = {
                "Admissão": {},
                "Demissão": {}
            }
            for grau in dicionarioDeDados["graudeinstrução"]:
                valor = dicionarioDeDados["graudeinstrução"][grau]
                dadosGrupos["Admissão"][valor] = 0
                dadosGrupos["Demissão"][valor] = 0
                    
            for chave, grupo in grupos:
                codigoTipoGrauInstrucao = int(pegarTipoColunaGrupo(grupo, "graudeinstrução"))
                codigoTipoMovimentacao =  int(pegarTipoColunaGrupo(grupo, "tipomovimentação"))

                total = pegarSomaTotalGrupo(grupo, "graudeinstrução")

                nomeTipo = dicionarioDeDados["graudeinstrução"][codigoTipoGrauInstrucao]
                nomeTipoMovimentacao = dicionarioDeDados["tipomovimentação"][codigoTipoMovimentacao]

                if nomeTipoMovimentacao == "Admissão":
                    dadosGrupos["Admissão"][nomeTipo] = total
                else:
                    dadosGrupos["Demissão"][nomeTipo] = total

            novosDados[municipio][mes] = dadosGrupos
        
    return novosDados
```

```python
dadosPorRegiao = organizarPorRegiao(dados)

dadosFiltrados = {}
for regiao in dadosPorRegiao:
    dadosFiltrados[regiao] = filtrarDados(dadosPorRegiao[regiao])
```

```python
dadosFiltradosEscolaridade = organizarDados(dadosPorRegiao)
```

```python
def plotarAdmissaoEscolaridade(dadosOrganizados, chave):
    for municipio in dadosOrganizados:
        if municipio == municipioDeConsulta:
            fig, ax = plt.subplots()
            meses = []
            rects = []
            dadosParaPlotar = {}
            
            for mes in dadosOrganizados[municipio]:
                meses.append(mes)

                for grauEscolaridade in dadosOrganizados[municipio][mes][chave]:
                    if grauEscolaridade not in dadosParaPlotar:
                        dadosParaPlotar[grauEscolaridade] = []
                    
                    dadosParaPlotar[grauEscolaridade].append(dadosOrganizados[municipio][mes][chave][grauEscolaridade])
            
            x = numpy.arange(len(meses))
            espacamento = -0.30
            area = 0.05
            for grauEscolaridade in dadosParaPlotar:
                rect = ax.bar(x + espacamento, dadosParaPlotar[grauEscolaridade], width=area, label=grauEscolaridade)
                espacamento += area + 0.01
                rects.append(rect)

            ax.set_ylabel(chave)
            ax.set_title('Meses')
            ax.set_xticks(x)
            ax.set_xticklabels(meses)
            ax.legend()

            for rect in rects:
                ax.bar_label(rect, padding=3)

            fig.tight_layout()

            fig.set_figwidth(18)
            fig.set_figheight(5)

            plt.grid(axis = 'y')

            plt.title(chave + " por grau de escolaridade para o munícipio de " + municipio + "\n" + anoDeConsulta + "\n")

            plt.show()
```

```python
def plotarAdmissao(dadosOrganizados):
    for municipio in dadosOrganizados:
        if municipio == municipioDeConsulta:
            fig, ax = plt.subplots()
            meses = []
            rects = []
            homens = []
            mulheres = []

            for mes in dadosOrganizados[municipio]:
                meses.append(mes)

                for genero in dadosOrganizados[municipio][mes]:
                    if genero == "Homem":
                        if "Admissão" not in dadosOrganizados[municipio][mes][genero]:
                            dadosOrganizados[municipio][mes][genero]["Admissão"] = 0

                        homens.append(dadosOrganizados[municipio][mes][genero]["Admissão"])
                    else:
                        if "Admissão" not in dadosOrganizados[municipio][mes][genero]:
                            dadosOrganizados[municipio][mes][genero]["Admissão"] = 0

                        mulheres.append(dadosOrganizados[municipio][mes][genero]["Admissão"])

            x = numpy.arange(len(meses))
            width = 0.25

            rect = ax.bar(x - width/2, homens, width, label='Homens')
            rect2 = ax.bar(x + width/2, mulheres, width, label='Mulheres')

            ax.set_ylabel('Admissões')
            ax.set_title('Meses')
            ax.set_xticks(x)
            ax.set_xticklabels(meses)
            ax.legend()

            ax.bar_label(rect, padding=3)
            ax.bar_label(rect2, padding=3)

            fig.tight_layout()

            fig.set_figwidth(18)
            fig.set_figheight(5)

            plt.grid(axis = 'y')

            plt.title("Admissões por gênero para o munícipio de " + municipio + "\n" + anoDeConsulta + "\n")

            plt.show()
```
